```python
data = []

random_baseline = ['random-baseline', 'n/a', '20', '80', '0.25']
human_gold_standard = ['human-gold-standard', 'n/a', 'n/a', '80', '0.8557'] # from moodle

data.append(random_baseline)
data.append(human_gold_standard)

with open("analysis.csv") as file:
    for line in file:
        data.append(line.split(","))
```

```python
import numpy as np
```

```python
data_np = np.array(data)
x = data_np[:,0]
y = data_np[:,4]
y = y.astype(float)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xticks(fontsize=60)
plt.yticks(fontsize=60)
plt.rcParams["figure.figsize"] = (100, 100)
plt.bar(x, y)
```
